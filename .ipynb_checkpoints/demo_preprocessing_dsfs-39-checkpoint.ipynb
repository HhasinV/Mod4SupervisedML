```python
df = pd.read_csv("00-Lectures/src/Dataset_preprocessing.csv")
```

```python
column_to_drop = df.isna().sum()[~((df.isna().sum() / df.shape[0]) < .7)].index
# df = df.drop(column_to_drop, axis=1)
```

```python
df.drop(column_to_drop, axis=1, inplace=True)
```

```python
df
```

```python
def super_complexe(x):
    return x.Age * x.Salary
```

```python
df["super_complexe"] = df.apply(super_complexe, axis=1)
```

```python
df.drop(["useless_col", "id", "super_complexe"], axis=1, inplace=True)
```

```python
from sklearn.impute import SimpleImputer

si = SimpleImputer()
df["Age"] = si.fit_transform(df[["Age", "Salary"]])
```

```python
to_transform = df[["Age", "Salary"]]
pd.DataFrame(si.fit_transform(to_transform), columns=to_transform.columns)
```

```python
df.isna().sum()
```

```python
def replace_age(age):
    if age < 0:
        return df[df["Age"] > 0]["Age"].mean()
    else : 
        return age

df["Age"] = df["Age"].apply(replace_age)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

X = df.drop("Purchased", axis=1)
y = df["Purchased"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=df["Country"])


numerical_columns = X_train.select_dtypes(exclude="object").columns
categorical_columns = X_train.select_dtypes(include="object").columns

numerical_pipeline = Pipeline([
    ("imputer", SimpleImputer()), 
    ("scale", StandardScaler())
    ])

categorical_pipeline = Pipeline([
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(drop="first"))
])

ct = ColumnTransformer(transformers=[
    ("numerical_transformers", numerical_pipeline, numerical_columns ),
    ("categorical_transformers", categorical_pipeline, categorical_columns)
])

X_train = ct.fit_transform(X_train) # Tout ce que j'apprend je l'apprend sur X_train. Puis j'applique au Xtrain
X_test = ct.transform(X_test) # Puis j'applique au Xtest

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```
